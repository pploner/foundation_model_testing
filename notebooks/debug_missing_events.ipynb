```python
import json

# Load JSON files once
with open("/afs/cern.ch/work/p/phploner/foundation_model_testing/src/utils/nEvents_scan/file_event_counts.json") as f:
    nevents = json.load(f)

with open("/eos/user/p/phploner/foundation_model_testing_data/v2_13class/vectorized/split_manifest_global.json") as f:
    splits = json.load(f)


def print_dataset_summary():
    print("\n================ DATASET SUMMARY ================\n")

    global_total = 0

    for sample in sorted(splits.keys()):
        if sample not in nevents:
            print(f"⚠️  {sample} not found in nevents.json\n")
            continue

        print(f"=== {sample} ===")

        sample_total = 0

        for split, files in splits[sample].items():
            total = 0
            missing = 0

            for fname in files:
                if fname in nevents[sample]:
                    total += nevents[sample][fname]
                else:
                    missing += 1

            sample_total += total
            global_total += total

            print(f"{split:>6} : {total:>12,} events   ({len(files)} files)", end="")
            if missing:
                print(f"   ⚠️ {missing} missing")
            else:
                print()

        print(f"{'TOTAL':>6} : {sample_total:>12,} events\n")

    print("===============================================")
    print(f"GLOBAL TOTAL: {global_total:,} events\n")
```

```python
print_dataset_summary()
```
